```python
import pandas as pd
import json
```

```python
with open('JSON_files/cwc_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

## Preprocessing

```python
df_match.rename({'scorecard':'matchid'}, inplace=True, axis=1)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['matchid']
    match_ids_dict[key2] = row['matchid']
```

```python
df_match.to_csv('CSV_files/cwc_results.csv', index = False)
```

## Batting data preprocessing

```python
with open('JSON_files/batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting['dismissal'].apply(lambda x: "Not Out" if pd.isna(x) else "Out")
df_batting.head(11)
```

```python
df_batting.drop(columns="dismissal", inplace=True)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('CSV_files/bating_summary.csv', index = False)
```

## Bowling data preprocessing

```python
with open('JSON_files/bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
  
df_bowling = pd.DataFrame(all_records)
df_bowling.head(11)
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('CSV_files/bowling_summary.csv', index = False)
```

## Player data preprocessing

```python
with open('JSON_files/players.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players[df_players['team'] == 'India']
```

```python
df_players.head()
```

```python
def standardize_name(name):
    return name.split('(')[0]

df_players['name'] = df_players['name'].apply(standardize_name)

# Now drop duplicates based on the standardized 'name' column
df_players.drop_duplicates(subset=['name'], keep='first', inplace=True)
df_players[df_players['team'] == 'India']
```

```python
df_players.to_csv('CSV_files/players.csv', index = False)
```
